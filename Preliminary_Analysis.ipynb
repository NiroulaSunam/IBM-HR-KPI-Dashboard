```python
#separate numerical and categorical columns 
def get_col_types(df):
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object', 'category','bool']).columns.tolist()
    datetime_cols = df.select_dtypes(include=['datetime','datetime64']).columns.tolist()
    return numeric_cols, categorical_cols,datetime_cols
```

```python
# Missing values 
#basic count
missing = df.isnull().sum()
display(missing[missing>0])

#percent missing 
missing_percent = missing/ df.shape[0]
display(missing_percent).sort_values(ascending=False).head(20)
```

```python
'''
if the data had missing values 
1. drop columns if 50-70% missing and not critical
2. drop rows if only a few rows missing crucical data
3. Impute 
Numeric -> median (robust)
           mean (if symmetric)
Categorical -> mode or "Missing" label
Use predictive imputation (KNN, iterative)

# numeric impute with median
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

# categorical impute with 'Missing'
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna('Missing', inplace=True)
'''
```

```python
#Bivariate analysis : feature vs target (Attrition)
# Numeric methods 1. Boxplots/violin plots (distribution by class)
# 2. Group means/ t-tests/ point-biserial correlation
# 3. Visual: OVerlay histograms for Attriton = 0 or 1 
for col in numeric_cols:
    if col == 'Attrition': continue
    plt.figure(figsize=(6,3))
    sns.boxplot(x='Attrition',y=col,data=df)
    plt.title(f"{col} by Attrtion")
    plt.show()
```

```python
from scipy.stats import ttest_ind
for col in numeric_cols:
    if col == 'Attrition': continue
    group0 = df[df['Attrition'] == 0][col].dropna()
    group1 = df[df['Attrition'] == 1][col].dropna()
    if len(group0) > 10 and len(group1) > 10:
        t, p = ttest_ind(group0, group1, equal_var=False)
        print(col, "t=", round(t,3), "p=", round(p,4))
```

```python
# t - size/direction of difference. a large +ve or -ve number means a big difference betn 2 groups
# p - tells whether the difference is statistically significant. p< 0.05 statistically significant difference. >= no significant difference
```

```python
## categorical methods : 1. crosstab with percentages
# 2. Chi- square test for independence
# 3. Stacked bar plots
for col in categorical_cols:
    ct = pd.crosstab(df[col], df['Attrition'])
    pct = ct.div(ct.sum(axis=1), axis=0)
    display(col, pct.round(3))
```

```python
from scipy.stats import chi2_contingency
for col in categorical_cols:
    ct = pd.crosstab(df[col], df['Attrition'])
    if ct.size > 1:
        chi2, p, dof, _ = chi2_contingency(ct)
        print(col, "p=", round(p,4))
```

```python
# Gives quick report or summary for analysis 
def quick_report(df, target='Attrition'):
    print("Shape:", df.shape)
    print("\nMissing values (%) top 10:")
    print((df.isnull().sum()/len(df)).sort_values(ascending=False).head(10))
    print("\nDtypes:")
    print(df.dtypes.value_counts())
    if target in df.columns:
        print("\nTarget distribution:")
        print(df[target].value_counts(normalize=True).round(3))
    print("\nTop numeric skew (abs) top 10:")
    num = df.select_dtypes(include=[np.number]).drop(columns=[target], errors='ignore')
    print(num.skew().abs().sort_values(ascending=False).head(10))
    print("\nColumns with one unique value:")
    print([c for c in df.columns if df[c].nunique()<=1])

quick_report(df, target='Attrition')
```
